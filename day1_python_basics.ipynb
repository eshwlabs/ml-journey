```python
# integer
print(4)
print(1e307) # means 10 to the power 307 (309 is max, gives infinity)
print(1e309)
```

```python
# float
print(1.5)
print(1.5e307)
```

```python
# boolean
print(True)
print(False)
```

**3. Variables**
- Dynamic typing


```python
 # in java, you declare the datatype of variable (called static typing), in python you dont (dynamic typing)
 # in java: String name="eshel";
name= "eshel"
print(name)
```

- Dynamic binding

```python
# when a variable can store multiple datatypes values
#unlike in java if we give int num, you can only store int values in it
name='eshel'
print(name)
name=4
print(name)
name=True
print(name)
```
